```python
# Phyton 3 program to for tree traversals
# Inorder
class Node:
    def __init__(self, key):
        self.left = None
        self.right = None
        self.val = key

def print_inorder(root):
    if root:
        # Recur on the left subtree
        print_inorder(root.left)
        # Print the value of the current node
        print(root.val, end=" ")
        # Recur on the right subtree
        print_inorder(root.right)

# Contoh pohon biner
root = Node(1)
root.left = Node(2)
root.right = Node(3)
root.left.left = Node(4)
root.left.right = Node(5)

# Cetak output "Inorder traversal of tree is 1 2 4 5 3"
print("Inorder traversal of tree is", end=" ")
print_inorder(root)

```
